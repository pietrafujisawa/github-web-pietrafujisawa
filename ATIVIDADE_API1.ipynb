# **DESENVOLVIMENTO 2**

```python
arquivoimp_1 = origem + 'IMP_2025.csv'
```

```python
arquivoimp_2 = origem + 'IMP_2026.csv'
```

```python
imp25 = pd.read_csv(arquivoimp_1,low_memory=False, sep=';',encoding='UTF-8')
imp26 = pd.read_csv(arquivoimp_2,low_memory=False, sep=';',encoding='UTF-8')
```

```python
imp_result = pd.concat([imp25,imp26], ignore_index=True)
imp_result.info()
```

```python
imp_result['DATA'] = pd.to_datetime(imp_result['CO_ANO'].astype(str) + '-' + imp_result['CO_MES'].astype(str) + '-01')
```

```python
grafico_dados = imp_result.groupby('DATA')['KG_LIQUIDO'].sum().reset_index()
```

```python
grafico_dados = grafico_dados[(grafico_dados['DATA'] >= '2025-01-01') & (grafico_dados['DATA'] <= '2026-01-01')]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(grafico_dados['DATA'], grafico_dados['KG_LIQUIDO'], marker='o', color='darkblue', linewidth=2)
plt.ticklabel_format(style='plain', axis='y')
plt.title('Evolução das Importações Brasileiras (Kg) - Jan/25 a Jan/26', fontsize=14)
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Total em Kilogramas (Kg)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
